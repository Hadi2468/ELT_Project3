```python
import pandas as pd
import glob
```

## date_dim

```python
# List all Parquet files in folder
files = glob.glob("../data/s3/date_dim/*.parquet")
display(files)

# Read and concatenate them
df_date_dim = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files], ignore_index=True)
display(df_date_dim.head())
df_date_dim.info()
```
