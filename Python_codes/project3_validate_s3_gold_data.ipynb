```python
import pandas as pd
import glob
```

# fact_customer_daily_metrics

```python
# List all Parquet files in folder
files = glob.glob("../data/s3/fact_customer/*.parquet")
display(files)

# Read and concatenate them
df_fact_customer = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files], ignore_index=True)
display(df_fact_customer.sort_values(by=['user_id', 'creation_date']).iloc[70:100,])
df_fact_customer.info()
```
