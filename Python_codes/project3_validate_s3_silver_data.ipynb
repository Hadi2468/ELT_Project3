# Project3: Validation of S3 Silver Layer

```python
import pandas as pd
import glob
```

## dim_date

```python
# List all Parquet files in folder
files = glob.glob("../data/s3/dim_date/*.parquet")
display(files)

# Read and concatenate them
df_dim_date = pd.concat([pd.read_parquet(f, engine="fastparquet") for f in files], ignore_index=True)
display(df_dim_date.sort_values(['date_key']).head().reset_index(drop=True))
df_dim_date.info()
```
